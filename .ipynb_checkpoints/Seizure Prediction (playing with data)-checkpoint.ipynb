```python
import scipy.io
import matplotlib.pyplot as plt

dataDir = '/Users/Gale/Documents/data';
featureDir = '/Users/Gale/Documents/Seizure-Prediction/features';

# Average and std of every channel for every training sample (in this case for the first patient)

average_1 = scipy.io.loadmat('/Users/Gale/Documents/Seizure-Prediction/features/average_train_1.mat');
std_1 = scipy.io.loadmat('/Users/Gale/Documents/Seizure-Prediction/features/std_train_1.mat');

Y_1 = scipy.io.loadmat('/Users/Gale/Documents/Seizure-Prediction/features/Y_1.mat');

numSamples = 720;

from pylab import rcParams
```

```python
average_1
```

```python
plt.plot(average_1['average'][0])
plt.plot(std_1['standardDev'][0])
plt.legend(['average', 'standardDev'], loc='upper left')
plt.show()
```

```python
avg_channel_0 = [] # channels from the interictal state (baseline)
avg_channel_1 = [] # channels from the preictal state (an hour prior to seizure)

for i in range(16):
    avg_channel_0.append([])
    avg_channel_1.append([])

for i in range(0, numSamples):
    if Y_1['Y'][i] == 0:            
        for j in range(0, 16):
            avg_channel_0[j].append(average_1['average'][i][j]) # ot std j-tiq channel na i-tiq sample
    else:
        for j in range(0, 16):
            avg_channel_1[j].append(average_1['average'][i][j])
            
# These are like plots on which the sample goes from left to right and the std is moving up and down
# and we'll have a plot for every channel, comparing interictal and preictal signals

plt.style.use('bmh') 
rcParams['figure.figsize'] = 6, 4

plt.hist(avg_channel_0[0])
plt.hist(avg_channel_1[0])
plt.legend(['interictal', 'preictal'], loc='upper left')
plt.show()
```

```python
# What about standard deviation?

std_channel_0 = [] # channels from the interictal state (baseline)
std_channel_1 = [] # channels from the preictal state (an hour prior to seizure)

# napr na vtoriq channel chetvurtoto
#print(std_1['standardDev'][3][1])

for i in range(16):
    std_channel_0.append([])
    std_channel_1.append([])

for i in range(0, numSamples):
    if Y_1['Y'][i] == 0:            
        for j in range(0, 16):
            std_channel_0[j].append(std_1['standardDev'][i][j]) # ot std j-tiq channel na i-tiq sample
    else:
        for j in range(0, 16):
            std_channel_1[j].append(std_1['standardDev'][i][j])
            
# These are like plots on which the sample goes from left to right and the std is moving up and down
# and we'll have a plot for every channel, comparing interictal and preictal signals


# Why are the current plots not good enough way of comparison? Because we are comparing average and std of preictal 
# and interictal stages and they are not something that can be comared like that with lines. Maybe we need points?
# maybe we need to plot it in one dimension or something like that where the height in which a point is the std/mean value
# and the y is the Y which is the type of state?

plt.style.use('bmh')

fig = plt.figure(figsize=(16, 7))

#subplot(nrows, ncols, plot_number)
one = fig.add_subplot(2, 2, 1)
one.plot(std_channel_0[0])
one.plot(std_channel_1[0])
one.legend(['interictal', 'preictal'], loc='upper left')
one.set_title("Standard Dev")

one = fig.add_subplot(2, 2, 2)
one.hist(std_channel_0[0])
one.hist(std_channel_1[0])
one.legend(['interictal', 'preictal'], loc='upper left')
one.set_title("Standard Dev")

one = fig.add_subplot(2, 2, 3)
one.plot(avg_channel_0[0])
one.plot(avg_channel_1[0])
one.legend(['interictal', 'preictal'], loc='upper left')
one.set_title("Average")

one = fig.add_subplot(2, 2, 4)
one.hist(avg_channel_0[0])
one.hist(avg_channel_1[0])
one.legend(['interictal', 'preictal'], loc='upper left')
one.set_title("Average")

plt.show()
```

```python
fig = plt.figure(figsize=(16, 7))

a = []
for i in range(len(std_channel_0[0])):
    a.append(0)
    
b = []
for i in range(len(std_channel_1[0])):
    b.append(1)

ax1 = fig.add_subplot(2, 1, 1)
ax1.scatter(std_channel_1[0], b, c = 'r')
ax1.scatter(std_channel_0[0], a, c = 'b')
ax1.legend(['preictal', 'interictal'], loc='upper left')
ax1.set_title("Standard Dev")

c = []
for i in range(len(avg_channel_0[0])):
    c.append(0)
    
d = []
for i in range(len(avg_channel_1[0])):
    d.append(1)
    
ax1 = fig.add_subplot(2, 1, 2)
ax1.scatter(avg_channel_1[0], d, c = 'r')
ax1.scatter(avg_channel_0[0], c, c = 'b')
ax1.legend(['preictal', 'interictal'], loc='upper left')
ax1.set_title("Average")

plt.show()
```
